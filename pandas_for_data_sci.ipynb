# คลิป 14 การสร้างกราฟเส้นแสดงการเปลี่ยนแปลงเป็น % เปรียบเทียบ

```python
df=pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/oil_gold_setindex.csv',
                index_col='Date', parse_dates=True)
df
```

```python
df.plot(kind='line')
```

```python
pd.set_option('display.float_format','{:,.5f}'.format)
```

```python
df['Oil_pct'] = df.Oil.pct_change()
df['Gold_pct'] = df.Gold.pct_change()
df['SET_pct'] = df.SET.pct_change()
df
```

```python
cols = ['Oil_pct', 'Gold_pct', 'SET_pct']
color = ['c','r', 'gold']
ax = df[cols].plot(kind='line')
ax.axhline(0)
ax.yaxis.set_major_formatter(FuncFormatter(lambda v,p :f'{v*100:.2f}%')) # ********************************
```

```python
cols = ['Oil_pct', 'Gold_pct', 'SET_pct']
color = ['c','r', 'gold']
ax = df[cols].plot(kind='line', subplots=True)
```

```python
cols = ['Oil_pct', 'Gold_pct', 'SET_pct']
color = ['c','r', 'gold']
ax = df[cols].plot(kind='line', sharey=True,subplots=True, layout=(1, 3), rot=90, figsize=(10, 3));   # rot คือ หมุน
# ax.yaxis.set_major_formatter(FuncFormatter(lambda v,p :f'{v:.2f}%')) ทำไม่ได้เหราะว่าเป็น subplots = True
```

```python
df=pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/oil_gold_setindex.csv',
                index_col='Date', parse_dates=True)
a = '2003-01-31'
df['Oil_diff'] = df.Oil/df.loc[a, 'Oil']-1
df['Gold_diff'] = df.Gold/df.loc[a, 'Gold']-1
df['SET_diff'] = df.SET/df.loc[a, 'SET']-1
df.head()
```

```python
ax = df.loc[:, df.columns[df.columns.str.contains('diff')]].plot(kind='line', grid=True)
ax.yaxis.set_major_formatter(FuncFormatter(lambda v,p :f'{v*100:.1f}%')) # ***************************************
```
